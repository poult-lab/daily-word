```python
with open('./static.txt','r') as f:
    filelist=f.readlines()

print("This the total line of file in txt: ",len(filelist))

new_list=[]

for x in filelist: 
    if x[0]=='#' and x[1]=='#' and x[2]=='#' and x[3]=='#':
        new_list.append(x)

number_counter=[]

for k in new_list:
    print(k)
    number_counter.append(k[-3:-1])
print(number_counter)

total=100
for m in number_counter:
    total=int(m)+total

print(total)
```
